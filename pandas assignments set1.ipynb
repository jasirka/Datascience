```python
stud_dataset = {'ID':[1,2,3,4], 'Name': ['Asin','Abi','Cini','Devi'],'Marks': [93,73,94,57]}
```

```python
df_stud = pd.DataFrame(stud_dataset)
```

```python
df_stud
```

```python
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 70 and marks <= 89:
        return 'B'
    else:
        return 'C'
```

```python
df_stud['Grade'] = df_stud['Marks'].apply(assign_grade)
```

```python
df_stud
```

```python
df_stud[(df_stud['Grade'] == 'A')]
```
