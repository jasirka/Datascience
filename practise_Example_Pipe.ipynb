```python
# Increase Monthly Income by 10%
def increase_income(df):
    df["UpdatedIncome"] = df["MonthlyIncome"] * 1.10
    return df

#How it works mathematically:

    #1.10 = 100% (original) + 10% (increase)

    #So, MonthlyIncome * 1.10 increases the salary by 10%

# Add a flag for Senior Employees (Age > 40)
def senior(df):
    df["IsSenior"] = df["Age"].apply(lambda x: "Yes" if x > 40 else "No")
    return df

# Assign grade based on Updated Income
def income_grade(df):
    df["IncomeGrade"] = df["UpdatedIncome"].apply(
        lambda x: "A" if x > 15000 else "B" if x > 7000 else "C"
    )
    return df
```

```python
df = (
    df.pipe(increase_income)
      .pipe(senior)
      .pipe(income_grade)
)

# Display first few rows
df[["Age", "MonthlyIncome", "UpdatedIncome", "IsSenior", "IncomeGrade"]].head()
```
