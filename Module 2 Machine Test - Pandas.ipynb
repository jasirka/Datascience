```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['Eby','Fisto','Clint','Binu','Sam'],
    'Age': [23,25,30,34,27],
    'Salary': [25000,15000,30000,40000,25000]
})
print(df)
```

```python
print(df.head(3))
```

```python
import numpy as np
from scipy import stats
print(f"Average Salary(Mean) = {np.mean(df['Salary'])}")
print(f"Highest Salary = {np.max(df['Salary'])}")
print(f"Smallest Salary = {np.min(df['Salary'])}")
print(f"Median Salary = {np.median(df['Salary'])}")
print(f"Mode = {stats.mode(df['Salary'])}")
print(f"Standard deviation = {np.std(df['Salary'])}")
print(f"Variance = {np.var(df['Salary'])}")
```
